```python
data1 = pd.read_csv(r'C:\Users\LENOVO\Desktop\Feature Engineering\mercedesbenz.csv', usecols = ['X1','X2','X3','X4','X5','X6'])
data1
```

```python
for col in data1.columns:
    print(col, ':', len(data1[col].unique()), 'labels')
```

```python
pd.get_dummies(data1, drop_first=True).shape
#after one hot encoding above 6 variables, total 117 Columns
```

```python
data1.X3.value_counts().sort_values(ascending=False).head(20)
# top 20 most frequent categories for the varibale X3
```

```python
data1.X2.value_counts().sort_values(ascending=False).head(20)
# top 20 most frequent categories for the varibale X2
```

```python
data1.X2.value_counts().sort_values(ascending=False).head(10)
# top 20 most frequent categories for the varibale X3
```

```python
top_10 = [x for x in data1.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10  #Index of top 10 most frequent categories for the variables x2
```

```python
for label in top_10:
    data1[label] = np.where(data1['X2']==label,1,0)
data1[['X2']+top_10].head(9)  #making 10 binary variables for x2
```

```python
def one_hot_top_x(df, variable, top_x_labels):
    for label in top_x_labels:
        df[variable+'_'+label] = np.where(data1[variable]==label,1,0)
        
one_hot_top_x(data1, 'X2', top_10)
data1.head()
```

```python
#finding top10 most frequent categories for x1
top_10 = [x for x in data1.X1.value_counts().sort_values(ascending=False).head(10).index]

#creating top dummy variables for x1
one_hot_top_x(data1, 'X1', top_10)
data1.head()
```
